```python
%config Completer.use_jedi = False
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('adult.data.csv')
data.head()
```

**1. How many men and women (*sex* feature) are represented in this dataset?** 

```python
# You code here
data['sex'].value_counts()
```

**2. What is the average age (*age* feature) of women?**

```python
# You code here
data.loc[data['sex'] == 'Female', ['age']].mean()
```

**3. What is the percentage of German citizens (*native-country* feature)?**

```python
# You code here
data['native-country'].value_counts(normalize=True)[['Germany']] * 100
```

**4-5. What are the mean and standard deviation of age for those who earn more than 50K per year (*salary* feature) and those who earn less than 50K per year?**

```python
# You code here
#average age for those who earn less than 50k per year
print('Salary types:', data['salary'].unique())
print('')
data.groupby(['salary'])['age'].agg([np.mean, np.std])
```

**6. Is it true that people who earn more than 50K have at least high school education? (*education – Bachelors, Prof-school, Assoc-acdm, Assoc-voc, Masters* or *Doctorate* feature)**

```python
HIGHER_EDUCATION = np.array(['Bachelors', 'Prof-school', 'Assoc-acdm', 'Assoc-voc', 'Masters', 'Doctorate'])
education_for_sal_more_than_50k = data.loc[data['salary'] == '>50K', 'education'].unique()
any_non_higher_education = np.setdiff1d(education_for_sal_more_than_50k, HIGHER_EDUCATION)
print('List of non high education for people who earn more than 50k:', any_non_higher_education)
print('Are all people who earn more than 50K have high school education:', len(any_non_higher_education) == 0)
```

**7. Display age statistics for each race (*race* feature) and each gender (*sex* feature). Use *groupby()* and *describe()*. Find the maximum age of men of *Amer-Indian-Eskimo* race.**

```python
# You code here
#show age statistics for race feature
sex_race_statistics = data.groupby(['sex','race'])['age'].describe()
display(sex_race_statistics)
amer_indian_men_max_age = sex_race_statistics.loc[('Male', 'Amer-Indian-Eskimo'), 'max']
print('')
print('Maximum age of men of Amer-Indian-Eskimo race:', amer_indian_men_max_age)
```

**8. Among whom is the proportion of those who earn a lot (>50K) greater: married or single men (*marital-status* feature)? Consider as married those who have a *marital-status* starting with *Married* (Married-civ-spouse, Married-spouse-absent or Married-AF-spouse), the rest are considered bachelors.**

```python
data.shape
```

```python
# You code here
print('Martial status types:', data['marital-status'].unique())
print('')
print('Martial status distribution for those who earn more than 50K:')
data['is_married'] = data['marital-status'].apply(lambda status: 'married' if status.startswith('Married') else 'bachelor' )
data.loc[data['salary'] == '>50K', 'is_married'].value_counts(normalize = True)
data.loc[(data['salary'] == '>50K') & (data['sex'] == 'Male'), 'is_married'].value_counts(normalize = True)
```

**9. What is the maximum number of hours a person works per week (*hours-per-week* feature)? How many people work such a number of hours, and what is the percentage of those who earn a lot (>50K) among them?**

```python
# You code here
max_hours = data['hours-per-week'].max()
total_people_amt = data.loc[data['hours-per-week'] == max_hours].shape[0]
pct_people_who_earn_a_lot = data.loc[data['hours-per-week'] == max_hours, 'salary'].\
                                value_counts(normalize = True)['>50K']
print('Maximum hours per week:', max_hours)
print('Amount of people who work maximum hours:', total_people_amt)
print(f'Pct of people with maximum hours and high sallary: {pct_people_who_earn_a_lot:.2%}')
```

**10. Count the average time of work (*hours-per-week*) for those who earn a little and a lot (*salary*) for each country (*native-country*). What will these be for Japan?**

```python
# You code here
avg_hours_per_week = data.pivot_table(index = ['native-country'], columns = ['salary'], values = ['hours-per-week'], aggfunc = 'mean')
display(avg_hours_per_week)
print('')
print('Japan average hours-per-week:')
display(avg_hours_per_week.loc[['Japan'],:])
```
